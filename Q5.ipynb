```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt

img=cv.imread('shells.tif',cv.IMREAD_GRAYSCALE)  #read  the image
hist=cv.calcHist([img],[0],None,[256],[0,256])  #histogram of the original image

table = np.array([i*2.8 for i in np.arange(0,256)]).astype('uint8')  #look up table of the transform
equ=cv.LUT(img,table)  #histogram of the equalized image
hist_equ=cv.calcHist([equ],[0],None,[256],[0,256])
cv.imwrite('Show.jpg',equ)

fig,ax =plt.subplots(1,2)
ax[0].plot(hist) 
ax[0].set_title('Original Histogram')
ax[0].set_xlim(0,255)
ax[1].plot(hist_equ)
ax[1].set_title('Equalized Histogram')
ax[1].set_xlim(0,255)
plt.savefig('Histogram')
plt.show()
```

```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt

img=cv.imread('shells.tif',cv.IMREAD_GRAYSCALE)  #read  the image
hist=cv.calcHist([img],[0],None,[256],[0,256])  #histogram of the original image

table = np.array([i*2.8 for i in np.arange(0,256)]).astype('uint8')  #look up table of the transform
equ=cv.LUT(img,table)  #histogram of the equalized image
hist_equ=cv.calcHist([equ],[0],None,[256],[0,256])

fig,ax=plt.subplots(1,2)
ax[0].imshow(img,cmap='gray',vmin=0,vmax=255)
ax[0].set_title('Original Image')
ax[1].imshow(equ,cmap='gray',vmin=0,vmax=255)
ax[1].set_title('Equalized Image')
plt.savefig('Compare')
plt.show()
```
