```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

#coordinates of break points
c=([50,50],[50,100],[150,255],[150,150])
#create the line segments
t1=np.linspace(0,c[0][1],c[0][0]-0+1).astype('uint8')
t2=np.linspace(c[1][1]+1,c[2][1],c[2][0]-c[1][0]).astype('uint8')
t3=np.linspace(c[3][1]+1,255,255-c[3][0]).astype('uint8')

#concatenate the linear pieces
transform=np.concatenate((t1,t2),axis=0).astype('uint8')
transform=np.concatenate((transform,t3),axis=0).astype('uint8')

#plot the transform
fig,ax=plt.subplots()
ax.plot(transform)
ax.set_xlabel('Input intensity')
ax.set_ylabel('Output intensity')
ax.set_xlim(0,256) #set the limit for the axis
ax.set_ylim(0,256)
ax.set_aspect('equal')
plt.savefig('Intensity transformation.jpg') #save the  image
plt.show()

img=cv.imread('emma.jpg',cv.IMREAD_GRAYSCALE)   #read the image as img
img_transformed=cv.LUT(img,transform) #do the transformation
cv.imwrite('emma_trnsformed.jpg',img_transformed)   #write the transformed image
```
