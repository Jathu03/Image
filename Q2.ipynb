```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

#create the line segments
x=180
t1=np.linspace(40,40,x-0+1).astype('uint8')
t2=np.linspace(255,255,255-x).astype('uint8')

#concatenate the linear pieces
transform=np.concatenate((t1,t2),axis=0).astype('uint8')
transform2=255-transform
#plot the transform
fig,ax=plt.subplots(1,2)
ax[0].plot(transform2)
ax[0].set_xlabel('Input intensity')
ax[0].set_ylabel('Output intensity')
ax[0].set_xlim(0,256) #set the limit for the axis
ax[0].set_ylim(0,256)
ax[0].set_aspect('equal')
#plt.savefig('Intensity transformation.jpg') #save the  image


img=cv.imread('BrainProtonDensitySlice9 (1).png',cv.IMREAD_GRAYSCALE) #read the image as img
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
img_transformed=cv.LUT(img,transform) #do the transformation
ax[1].imshow(img_transformed)

plt.show()
```
